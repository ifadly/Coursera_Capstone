## **Severity classification of traffic accidents in the UK**

## **Introduction**

A traffic accident occurs when a vehicle collides with another vehicle, pedestrian, animal, road debris, or other stationary obstruction, such as a tree, pole or building. Traffic collisions often result in injury, disability, death, and property damage as well as financial costs to both society and the individuals involved. In 2013, traffic accidents resulted in 1.4 million fatalities and 54 million injuries. [[Wikipedia](https://en.wikipedia.org/wiki/Traffic_collision)]


Traffic accidents may be caused by a long list of factors including weather, mechanical problems, road conditions, driver state of mind, road design. Proper reporting, recording and analysis of traffic accidents data will be helpful for urban planners, car designers and drivers to paint a more detailed picture on where, when and how accidents happen, leading to reductions in the total number of traffic accidents and hopfuly elimination of fatal accidents all together.
This report is intended to prsent the results of traffic accidents data analysis and ML modelling with the aim of predicting the severity of accidents based on a number of factors.


# **Data**

This analysis is based on data made available by the government of UK, the [dataset](https://www.kaggle.com/tsiaras/uk-road-safety-accidents-and-vehicles) lists over 2 million recoeds of accidents in the time frame between 2005 and 2017 within the geographical boundries of UK.
a total of 34 attributes details each of the accidents records, the detailed attributes provided for each accident include accident severity, dates, days of the week, light and weather conditions, road surface conditions, locations, etc.
See below figures 1 to 3 for an overview of the dataset used in this project.

Based on data exploration, it is evident that the number of missing attributes ("unknown" or "data missing or out of range") is minor, see fig.3 for sample,  and can be removed without impacting the integrity of the dataset.

The following attributes will be considered as contributing factors after removing missing data and normalization;

* Hazard existing on the road & any Special conditions
* Day of the week
* Road junction type
* Light conditions
* Weather Conditions
* Road surface condition
* Road type

The above attributes will be used to build a machine learning model to predict the severity of a traffic accidents. 

Below are some observations on the selected attributes;
* It was noticed that "Hazards" are in some cases indicated as "none" while special conditions indicate present hazards, such as "oil/diesel spell" for example. Hence the decision to merge both attributes in one.
* The number of accidents vary depending of weekdays, the highest number of accidents, 335183, took place on Fridays and the lowest being 225327 on Sundays.
* The data shows that around 40% of accidents happen with no junction/crossing within 20 meters, however, 30% took place at T-junctions.
* Surprisignly, 80% of accidents took place on days with fine weather.
* Almost 70% of accidents took place on dry roads and approximatly 28% on wet/damp roads
* Almost 75% of accidents took place on single carriageway (undevided roads)
